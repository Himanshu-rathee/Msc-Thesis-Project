```python
# Configure
c = twint.Config()
c.Lang = "en"
#c.Username = "newsinvesting"
c.Limit = 2000
c.Since = '2011-01-04'
c.until= '2021-06-25'
c.Pandas = True
c.Output = "tweets.json"
c.Search = ['India','Economy','Policy','RBI']

# Run
twint.run.Search(c)
Tweets_df = twint.storage.panda.Tweets_df
```
