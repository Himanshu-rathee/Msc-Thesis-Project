```python
import codecs, json, csv
import pandas as pd
```

```python
from textblob import TextBlob

MixedTweet = pd.read_csv('Tweets3.csv',index_col=0)
```

```python
MixedTweet
```

```python
sentyEng = []

for i,text in enumerate(MixedTweet.tweet):
  blob = TextBlob(text)
  print(blob.sentiment)
  if blob.sentiment[0]>0:
     print('Positive')
     sentyEng.append('Positive')
  elif blob.sentiment[0]<0:
     print('Negative')
     sentyEng.append('Negative')
  else:
     print('Neutral')
     sentyEng.append('Neutral')
```

```python
MixedTweet['Sentiments'] = sentyEng
MixedTweet.to_csv('C:/Users/rathe/Thesis Code/Sentitweets.csv')
```
